# 01- The single neuron
> **Goal:** Understand what one neuron does, step by step.
A neuron does exactly 3 things:
1. Multiply each input by a weight
2. Add a bias
3. Apply an activation function

> output = f(w.x +b)


```python
import numpy as np
#Activation function

def sigmoid(x):
  return 1/(1+np.exp(-x))

def relu(x):
  return np.maximum(0,x)

def neuron(x,w,b, activation=sigmoid):
  z=np.dot(w,x)+b
  return activation(z)
```

```python
# for instance: rain prediction
# x1 = humidity (0 to 1)
# x2 = cloud cover(0 to 1)
# x3 = wind speed (0 to 1)

x = np.array([0.9,0.7,0.3])
w = np.array([0.8,0.5,0.2])
b = 0.1

output = neuron(x,w,b,sigmoid)
print(output)
```
